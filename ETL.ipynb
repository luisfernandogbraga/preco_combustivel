```python

import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('anp.csv')
display(df)
```

```python
df.info()
```

```python
df['Data da Coleta'] = pd.to_datetime(df['Data da Coleta'], format="%d/%m/%Y")
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df= df.dropna()
display(df)
```

```python
for coluna in df.columns:
    print(f'Valores únicos em {coluna}: {df[coluna].unique()}')
```

```python
df.columns= df.columns.str.replace('Produto','Combustível')
```

```python
display(df)
```

```python
plt.figure(figsize=(18, 6))
sns.countplot(x='Combustível', data=df, palette='viridis')
plt.grid(True,linestyle='--', alpha=0.7)
plt.title('Combustíveis mais usados')
plt.xlabel('Combustível')
plt.ylabel('Contagem')
plt.show()
```

Gasolina de fato é o combustível mais usado, seguido pelo Etanol e o Gás Natural é o menos usado.

```python
plt.figure(figsize=(18, 8))
sns.barplot(x='Combustível', y='Valor de Venda', data=df, palette='viridis', ci=None)
plt.grid(True,linestyle='--', alpha=0.7)
plt.title('Média de Valor de Venda por Combustível')
plt.xlabel('Combustível')
plt.ylabel('Média de Valor de Venda')
plt.show()
```

```python
# Solicitar o nome do produto usando input
combustivel = input('Digite o nome do produto para visualizar o histórico de preços: ')

# Filtrar o DataFrame para o produto escolhido
df_produto = df[df['Combustível'] == combustivel]

# Verificar se há dados para o produto
if df_produto.empty:
    print(f'Não há dados disponíveis para o produto "{combustivel}".')
else:
    # Calcular valores estatísticos
    media_valor = df_produto['Valor de Venda'].mean()
    max_valor = df_produto['Valor de Venda'].max()
    min_valor = df_produto['Valor de Venda'].min()

    # Criar um gráfico de linha para o produto escolhido
    plt.figure(figsize=(18, 6))
    sns.lineplot(x='Data da Coleta', y='Valor de Venda', data=df_produto, ci=None)
    
    plt.grid(True, linestyle='--', alpha=0.7)

    # Adicionar linhas para média, máximo e mínimo
    plt.axhline(y=media_valor, color='r', linestyle='--', label='Média')
    plt.text(df_produto['Data da Coleta'].max(), media_valor + 0.1, f'Média: {media_valor:.2f}', color='r', ha='left', va='center')

    # Adicionar título e rótulos aos eixos
    plt.title(f'Histórico de Preços do Combustível: {combustivel}')
    plt.xlabel('Data da Coleta')
    plt.ylabel('Valor de Venda')

     # Adicionar legenda
    plt.legend()

    # Mostrar o gráfico
    plt.show()
```

```python
nome_combustivel= input('Digite o nome do produto para visualizar o histórico de preços: ')
df_combustivel = df[df['Combustível'] == nome_combustivel]

# Configurar o tamanho da figura
plt.figure(figsize=(18, 8))

# Criar o boxplot
sns.boxplot(x='Estado - Sigla', y='Valor de Venda', data=df_combustivel, palette='Set2')


plt.grid(True, linestyle='--', alpha=0.7)
plt.title(f'Boxplot de Valores de Venda de {nome_combustivel} por Região')
plt.xlabel('Região')
plt.ylabel('Valor de Venda')


plt.show()
```

```python
# Solicitar ao usuário que digite o nome do produto
nome1_combustivel = input('Digite o nome do produto para visualizar o histórico de preços: ')

# Filtrar o DataFrame para o combustível selecionado
df_combustivel = df[df['Combustível'] == nome1_combustivel]

media_valor_por_estado = df_combustivel.groupby('Estado - Sigla')['Valor de Venda'].mean().sort_values()

# Configurar o tamanho da figura
plt.figure(figsize=(18, 8))

    # Criar o gráfico de barras
plt.bar(media_valor_por_estado.index, media_valor_por_estado.values, color=sns.color_palette('Set1'))

    # Adicionar detalhes ao gráfico
plt.grid(True, linestyle='--', alpha=0.7)
plt.title(f'Média de Valor de Venda de {nome1_combustivel} por Estado')
plt.xlabel('Estado')
plt.ylabel('Média de Valor de Venda')

   
plt.show()
```

```python
# Solicitar ao usuário que digite o nome do produto
nome1_combustivel = input('Digite o nome do produto para visualizar o histórico de preços: ')

# Filtrar o DataFrame para o combustível selecionado
df_combustivel = df[df['Combustível'] == nome1_combustivel]

media_valor_por_estado = df_combustivel.groupby('Regiao - Sigla')['Valor de Venda'].mean().sort_values()

# Configurar o tamanho da figura
plt.figure(figsize=(18, 8))

    # Criar o gráfico de barras
plt.bar(media_valor_por_estado.index, media_valor_por_estado.values, color=sns.color_palette('Set1'))

    # Adicionar detalhes ao gráfico
plt.grid(True, linestyle='--', alpha=0.7)
plt.title(f'Média de Valor de Venda de {nome1_combustivel} por Região')
plt.xlabel('Região')
plt.ylabel('Média de Valor de Venda')

   
plt.show()
```

```python
# REGIOES : SE , S , CO , NE , N 

# Solicitar ao usuário que digite o nome do produto
nome2_combustivel = input('Digite o nome do combustível para visualizar o histórico de preços: ')
nome_regiao = input('Digite o nome da região para filtrar os dados: ')

# Filtrar o DataFrame para o combustível e região selecionados
df_filtrado = df[(df['Combustível'] == nome2_combustivel) & (df['Regiao - Sigla'] == nome_regiao)]

media_valor_por_estado = df_filtrado.groupby('Estado - Sigla')['Valor de Venda'].mean().sort_values()

# Configurar o tamanho da figura
plt.figure(figsize=(18, 8))

    # Criar o gráfico de barras
plt.bar(media_valor_por_estado.index, media_valor_por_estado.values, color=sns.color_palette('Set1'))

    # Adicionar detalhes ao gráfico
plt.grid(True, linestyle='--', alpha=0.7)
plt.title(f'Média de Valor de Venda de {nome2_combustivel} na Região {nome_regiao}')
plt.xlabel('Estado')
plt.ylabel('Média de Valor de Venda')

    # Mostrar o gráfico
plt.show()

```
